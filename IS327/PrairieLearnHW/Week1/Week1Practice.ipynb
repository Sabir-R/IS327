```python
import pandas as pd
```

```python
df = pd.read_csv("download.csv")
```

```python
df
```

```python
df["Sales"].mean()
```

```python
# Task 1: Impute Missing Values
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
df["Ratings"] = df["Ratings"].fillna(df["Ratings"].mean())
df["Product Category"] = df["Product Category"].fillna(df["Product Category"].value_counts().index[0])
imputed_products_df = df
imputed_products_df
```

```python
# Task 2: Filter Products
electronics_products = df.loc[df["Product Category"] == "Electronics"]
electronics_products
```

```python
# Task 3: Advanced Filter
filtered_electronics = df.loc[(df["Product Category"] == "Electronics") & (df["Sales"] > 499) & (df["Sales"] < 1001)] 
filtered_electronics
```

```python
# Task 4: Find Top Seller
topindex = df["Product ID"].index[df["Sales"].idxmax()]
top_seller_id = df["Product ID"].at[topindex]
top_seller_id
```

```python
# Task 5: Sales Threshold
df2 = df
df2["High Sales"] = (df2["Sales"] > 500).astype(int)
imputed_products_with_high_sales_df = df2
imputed_products_with_high_sales_df
```
